## Example: LTI (linear time-invariant) system

Plotting system's impulse response:

```python
from volterra import *
import matplotlib.pyplot as plt


sys_impulse_response = [3, 2, 1, 0.5, 0.1]
sys = LTISystem(sys_impulse_response)

x = [1] + [0] * 10
y = sys.evaluate_output(x)

plt.plot(y)
plt.grid(True)
plt.show()
```

Plotting system's step response:

```python
x = [1] * 10
y = sys.evaluate_output(x)

plt.plot(y)
plt.grid(True)
plt.show()
```

Plotting system's initial conditions response:

```python
x = [0] * 10
x0 = [4, 3, 2, 1]  # initial values at t = -4, t = -3, t = -2, t = -1
y = sys.evaluate_output(x, x0=x0)

plt.plot(y)
plt.grid(True)
plt.show()
```
